```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear el grafo del recinto
G = nx.Graph()

# Añadir nodos (lugares del concierto)
lugares = ["Entrada", "Pasillo1", "Pasillo2", "Escenario", "Baños", "Comida", "Salida"]
G.add_nodes_from(lugares)

# Conectar nodos con caminos
caminos = [("Entrada", "Pasillo1"), ("Pasillo1", "Escenario"), ("Pasillo1", "Baños"),
           ("Pasillo1", "Comida"), ("Pasillo1", "Pasillo2"), ("Pasillo2", "Escenario"),
           ("Pasillo2", "Salida")]

G.add_edges_from(caminos)

# Dibujar el grafo del recinto
plt.figure(figsize=(8,6))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=2000, edge_color='gray', font_size=12)
plt.show()
```

```python
import random

def mover_persona(persona, G):
    if persona["ubicacion"] == "Salida":
        return  # La persona ya salió del concierto

    # Seleccionar un destino aleatorio entre los vecinos
    posibles_movimientos = list(G.neighbors(persona["ubicacion"]))
    persona["ubicacion"] = random.choice(posibles_movimientos)

# Definir una persona en el modelo
persona = {"ubicacion": "Entrada"}

# Simular movimiento de la persona en varias iteraciones
for i in range(10):
    mover_persona(persona, G)
    print(f"Paso {i+1}: Persona está en {persona['ubicacion']}")
```

```python
import matplotlib.animation as animation

fig, ax = plt.subplots(figsize=(8,6))

# Posiciones fijas para los nodos
posiciones = nx.spring_layout(G)

global ani  # Mantiene la referencia para evitar que se elimine

def actualizar(i):
    ax.clear()
    mover_persona(persona, G)
    
    nx.draw(G, pos=posiciones, with_labels=True, node_color='skyblue', node_size=2000, edge_color='gray', font_size=12)
    
    # Dibujar a la persona en su ubicación actual
    nx.draw_networkx_nodes(G, pos=posiciones, nodelist=[persona["ubicacion"]], node_color='red', node_size=2500)

ani = animation.FuncAnimation(fig, actualizar, frames=20, repeat=False)
plt.show()  # Muestra la animación
```

```python
import random
from mesa import Agent, Model
from mesa.space import MultiGrid
from mesa.datacollection import DataCollector
from mesa.agent import AgentSet  # Nueva forma de manejar agentes
import matplotlib.pyplot as plt

class Person(Agent):
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.state = "Sano"  # Puede ser "Sano", "Infectado", o "Recuperado"
        self.infection_duration = 0

    def move(self):
        possible_steps = self.model.grid.get_neighborhood(
            self.pos,
            moore=True,
            include_center=False)
        new_position = self.random.choice(possible_steps)  # Usar self.random
        self.model.grid.move_agent(self, new_position)

    def infect(self):
        if self.state == "Sano":
            cellmates = self.model.grid.get_cell_list_contents([self.pos])
            if len(cellmates) > 1:
                for other in cellmates:
                    if other.state == "Infectado" and self.random.random() < self.model.infection_prob:
                        self.state = "Infectado"
                        self.infection_duration = self.model.recovery_time

    def recover(self):
        if self.state == "Infectado":
            self.infection_duration -= 1
            if self.infection_duration <= 0:
                self.state = "Recuperado"

    def step(self):
        self.move()
        self.infect()
        self.recover()

class ConcertModel(Model):
    def __init__(self, N, width, height, infection_prob, recovery_time):
        self.num_agents = N
        self.grid = MultiGrid(width, height, True)
        self.infection_prob = infection_prob
        self.recovery_time = recovery_time
        self.random = random.Random()  # Generador de números aleatorios
        self.schedule = AgentSet(4)  # Usar AgentSet en lugar de RandomActivation

        for i in range(self.num_agents):
            a = Person(i, self)
            self.schedule.add(a)
            x = self.random.randrange(self.grid.width)
            y = self.random.randrange(self.grid.height)
            self.grid.place_agent(a, (x, y))

        self.datacollector = DataCollector(
            agent_reporters={"State": "state"})

    def step(self):
        self.datacollector.collect(self)
        for agent in self.schedule:
            agent.step()

def get_infected_count(model):
    return sum(1 for a in model.schedule.agents if a.state == "Infectado")

def animate_model(model, steps):
    infected_counts = []
    for i in range(steps):
        model.step()
        infected_counts.append(get_infected_count(model))
        plt.clf()
        plt.grid(True)
        plt.title(f"Step {i+1}")
        plt.xlabel("X Coordinate")
        plt.ylabel("Y Coordinate")
        for agent in model.schedule.agents:
            if agent.state == "Sano":
                color = 'blue'
            elif agent.state == "Infectado":
                color = 'red'
            else:
                color = 'green'
            plt.scatter(agent.pos[0], agent.pos[1], color=color)
        plt.pause(0.1)
    plt.show()

    plt.plot(infected_counts)
    plt.title("Número de infectados a lo largo del tiempo")
    plt.xlabel("Step")
    plt.ylabel("Número de infectados")
    plt.show()

# Parámetros del modelo
N = 50  # Número de agentes
width = 20  # Ancho del grid
height = 20  # Alto del grid
infection_prob = 0.1  # Probabilidad de contagio
recovery_time = 10  # Tiempo de recuperación

# Crear y animar el modelo
model = ConcertModel(N, width, height, infection_prob, recovery_time)
animate_model(model, 100)
```
